```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fold_0 = pd.read_json("fold_0_modified/results/metrics.json", orient="records", lines=True)
fold_0 = fold_0.sort_values("iteration")
fold_0['fold'] = 0
fold_0
```

```python
fold_1 = pd.read_json("fold_1_modified/results/metrics.json", orient="records", lines=True)
fold_1 = fold_1.sort_values("iteration")
fold_1['fold'] = 1
fold_1
```

```python
fold_2 = pd.read_json("fold_2_modified/results/metrics.json", orient="records", lines=True)
fold_2 = fold_2.sort_values("iteration")
fold_2['fold'] = 2
fold_2
```

```python
fold_3 = pd.read_json("fold_3_modified/results/metrics.json", orient="records", lines=True)
fold_3 = fold_3.sort_values("iteration")
fold_3['fold'] = 3
fold_3
```

```python
fold_4 = pd.read_json("fold_4_modified/results/metrics.json", orient="records", lines=True)
fold_4 = fold_4.sort_values("iteration")
fold_4['fold'] = 4
fold_4
```

```python
fold = pd.concat([fold_0, fold_1, fold_2, fold_3, fold_4], axis=1)
```

```python
fold.index
```

```python
draw_df = pd.DataFrame(index=fold.index, columns=
                       [
                           'train_loss_min',
                           'train_loss_mean',
                           'train_loss_max',
                           
                           'validation_loss_min',
                           'validation_loss_mean',
                           'validation_loss_max',
                           
                           'validation_bbox/AP50_min',
                           'validation_bbox/AP50_mean',
                           'validation_bbox/AP50_max',
                           
                       ]
                      
                      )
```

```python
draw_df
```

```python
draw_df['train_loss_min'] = fold['total_loss'].min(axis=1)
draw_df['train_loss_mean'] = fold['total_loss'].mean(axis=1)
draw_df['train_loss_max'] = fold['total_loss'].max(axis=1)

draw_df['validation_loss_min'] = fold['validation_loss'].min(axis=1)
draw_df['validation_loss_mean'] = fold['validation_loss'].mean(axis=1)
draw_df['validation_loss_max'] = fold['validation_loss'].max(axis=1)

draw_df['validation_bbox/AP50_min'] = fold['bbox/AP50'].min(axis=1)
draw_df['validation_bbox/AP50_mean'] = fold['bbox/AP50'].mean(axis=1)
draw_df['validation_bbox/AP50_max'] = fold['bbox/AP50'].max(axis=1)
```

```python
draw_df_train_loss = draw_df[~draw_df["train_loss_mean"].isna()][['train_loss_min', 'train_loss_mean','train_loss_max']]
draw_df_validation_loss = draw_df[~draw_df["validation_loss_mean"].isna()][['validation_loss_min','validation_loss_mean','validation_loss_max']]
draw_df_ap = draw_df[~draw_df["validation_bbox/AP50_mean"].isna()][['validation_bbox/AP50_min','validation_bbox/AP50_mean','validation_bbox/AP50_max']]
```
