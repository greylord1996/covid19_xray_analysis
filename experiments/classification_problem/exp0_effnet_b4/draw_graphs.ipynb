```python
import numpy as np

import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
plt.style.use(['bmh'])
matplotlib.rcParams['axes.titlesize'] = 44
matplotlib.rcParams['axes.labelsize'] = 32
matplotlib.rcParams['axes.linewidth'] = 2
matplotlib.rcParams['lines.linewidth'] = 4
matplotlib.rcParams['lines.markersize'] = 10
matplotlib.rcParams['xtick.labelsize'] = 32
matplotlib.rcParams['ytick.labelsize'] = 32
matplotlib.rcParams['legend.fontsize'] = 32
matplotlib.rcParams['legend.frameon'] = False
# matplotlib.rcParams['font.family'] = 'serif'
matplotlib.rcParams['font.size'] = 32
matplotlib.rcParams['mathtext.default'] = 'regular'
matplotlib.rcParams['grid.linestyle'] = 'dashed'



def plot_train_valid(train, valid, ylabel, filename):
#     assert train.shape == valid.shape == (5, 18)
    # assert ylabel in ['Dice Coefficient', 'Loss']

    fig, ax = plt.subplots(figsize=(16, 10))
    color_train = '#1565C0' # if ylabel == 'Loss' else '#7E57C2'
    color_valid = '#EF6C00' # if ylabel == 'Loss' else '#C62828'

    # Set axis ranges; by default this will put major ticks every 25.
    # ax.set_xlim(1, 20)
    # if ylabel == 'Loss':
    #     ax.set_ylim(0.0, 1.6)
    # else:
    #     ax.set_ylim(0.15, 0.95)

    # Change major ticks to show every 20.
    ax.xaxis.set_major_locator(MultipleLocator(2))
    # ax.yaxis.set_major_locator(MultipleLocator(0.15))

    # Change minor ticks to show every 5. (20/4 = 5)
    ax.xaxis.set_minor_locator(AutoMinorLocator(2))
    # ax.yaxis.set_minor_locator(AutoMinorLocator(2))

    # Turn grid on for both major and minor ticks and style minor slightly
    # differently.
    ax.grid(which='major', color='#CCCCCC')
    ax.grid(which='minor', color='#CCCCCC')

    # plt.title("UnEfficientNet-B4 Training")
    plt.xlabel("Epochs")
    plt.ylabel(ylabel)

    x = np.array(range(1, train.shape[1]+1))

    plt.plot(x, train.mean(axis=0), label='train', color=color_train)
    plt.fill_between(
        x, y1=train.min(axis=0), y2=train.max(axis=0),
        alpha=0.5, color=color_train
    )

    plt.plot(x, valid.mean(axis=0), label='validation', color=color_valid)
    plt.fill_between(
        x, y1=valid.min(axis=0), y2=valid.max(axis=0),
        alpha=0.4, color=color_valid
    )

    plt.grid()
    plt.legend(loc=('upper right' if ylabel == 'Loss' else 'lower right'))
    plt.tight_layout()
    plt.savefig(filename + '.pdf', dpi=300, format='pdf')
```

```python
plot_train_valid(
    train=fold['average_precision_train'].T * 3/2,
    valid=fold['average_precision_valid'].T * 3/2,
    ylabel='AP',
    filename='results/EffNet-B4_average_precision_real'
)
```
