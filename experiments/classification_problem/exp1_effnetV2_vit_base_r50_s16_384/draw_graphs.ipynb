```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fold_0 = pd.read_csv("weights/vit_base_r50_s16_384_fold0_final.csv")
fold_0 = fold_0.loc[:fold_0.loc[[fold_0.average_precision_valid.idxmax()]].values[0][0]]
fold_0
```

```python
fold_1 = pd.read_csv("weights/vit_base_r50_s16_384_fold1_final.csv")
fold_1 = fold_1.loc[:fold_1.loc[[fold_1.average_precision_valid.idxmax()]].values[0][0]]
fold_1
```

```python
fold_2 = pd.read_csv("weights/vit_base_r50_s16_384_fold2_final.csv")
fold_2 = fold_2.loc[:fold_2.loc[[fold_2.average_precision_valid.idxmax()]].values[0][0]]
fold_2
```

```python
fold_3 = pd.read_csv("weights/vit_base_r50_s16_384_fold3_final.csv")
fold_3 = fold_3.loc[:fold_3.loc[[fold_3.average_precision_valid.idxmax()]].values[0][0]]
fold_3
```

```python
fold_4 = pd.read_csv("weights/vit_base_r50_s16_384_fold4_final.csv")
fold_4 = fold_4.loc[:fold_4.loc[[fold_4.average_precision_valid.idxmax()]].values[0][0]]
fold_4
```

```python
fold = pd.concat([fold_0, fold_1, fold_2, fold_3, fold_4], axis=1)
```
